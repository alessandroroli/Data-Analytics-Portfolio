```python
import pandas as pd
```

```python
# Carico il CSV file
df_new = pd.read_csv('D:\Lorenzo Viaro\Corso Boolean\Corso Data Analytics\FinalProject\Data\Raw\Product_performance.csv')


df_new.head()
```

```python
# Convert 'order_item_created_at' to datetime and keep only the date
df_new['order_item_created_at'] = pd.to_datetime(df_new['order_item_created_at']).dt.date
```

```python
# Drop rows where 'product_name' is missing
df_new_cleaned = df_new.dropna(subset=['product_name'])
```

```python
# Remove duplicates
df_new_cleaned = df_new_cleaned.drop_duplicates()
```

```python
# Check consistency of 'total_profit'
df_new_cleaned['calculated_profit'] = df_new_cleaned['product_retail_price'] - df_new_cleaned['product_cost']
profit_inconsistencies = df_new_cleaned[abs(df_new_cleaned['total_profit'] - df_new_cleaned['calculated_profit']) > 0.01]
```

```python
# Replace 'total_profit' with 'calculated_profit' for inconsistent rows
df_new_cleaned.loc[profit_inconsistencies.index, 'total_profit'] = df_new_cleaned.loc[profit_inconsistencies.index, 'calculated_profit']
```

```python
# Drop the 'calculated_profit' column as it's no longer needed
df_new_cleaned = df_new_cleaned.drop(columns='calculated_profit')
df_new_cleaned.head()
```

```python
# Check consistency again
df_new_cleaned['calculated_profit'] = df_new_cleaned['product_retail_price'] - df_new_cleaned['product_cost']
profit_inconsistencies_after = df_new_cleaned[abs(df_new_cleaned['total_profit'] - df_new_cleaned['calculated_profit']) > 0.01]
df_new_cleaned = df_new_cleaned.drop(columns='calculated_profit')
```

```python
#Check user_country column unique values
print(df_new_cleaned['user_country'].unique())
```

```python
# Replace "España" with "Spain" and "Brasil" with "Brazil"
df_new_cleaned['user_country'] = df_new_cleaned['user_country'].replace({"España": "Spain", "Brasil": "Brazil"})
```

```python
# Check the 'product_brand' column for anomalies
unique_brands = df_new_cleaned['product_brand'].value_counts()
unique_brands
```

```python
# Save the cleaned data to a new CSV file
df_new_cleaned.to_csv('Product_performance_cleaned.csv', index=False)
```
